```python
# Importing the required libraries
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the audio file
audio_path = 'harvard.wav'
audio_data, sample_rate = librosa.load(audio_path)
```

```python
# Calculate the duration of the audio
duration = librosa.get_duration(y=audio_data, sr=sample_rate)
print(f"Audio duration: {duration} seconds")
```

```python
# Visualize the waveform
plt.figure(figsize=(12, 4))
plt.plot(audio_data)
plt.title('Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Extract Mel spectrogram features
mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sample_rate, n_mels=128)
```

```python
# Visualize the Mel spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(librosa.power_to_db(mel_spectrogram, ref=np.max),
                         x_axis='time', y_axis='mel', sr=sample_rate, fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.tight_layout()
plt.show()
```

```python

# Extract MFCC features
mfcc = librosa.feature.mfcc(y=audio_data, sr=sample_rate, n_mfcc=13)
```

```python
# Visualize the MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

```python
# Extract chromagram features
chromagram = librosa.feature.chroma_stft(y=audio_data, sr=sample_rate)
```

```python
# Visualize the chromagram
plt.figure(figsize=(10, 4))
librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', cmap='coolwarm')
plt.colorbar()
plt.title('Chromagram')
plt.tight_layout()
plt.show()
```
